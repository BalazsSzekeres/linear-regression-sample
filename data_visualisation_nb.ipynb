```python
import pandas
from pandas import DataFrame
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pandas.read_csv('data/cost_revenue_clean.csv')

fig, ax = plt.subplots()

regression = LinearRegression()
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
regression.fit(X=X, y=y)
plt.plot(X, regression.predict(X), color='red')


data.plot.scatter(x='production_budget_usd', y='worldwide_gross_usd',
                  title="Film cost versus Revenue",
                  xlabel="Production budget", ylabel="Worldwide Gross",
                  ylim=[0,3000000000], xlim=[0,450000000],
                  alpha=0.3,
                  ax=ax)
plt.show()


```
